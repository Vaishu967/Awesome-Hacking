   **4.1 Sales analysis**

```python
##Total sales of each Region
df.groupby("Region")["Sales"].sum()
```

```python
##sales analysis based on region
df.groupby("Region")["Sales"].sum().plot.bar()
```

```python
##Total sales of each segment
df.groupby("Segment")["Sales"].sum()
```

```python
##sales analysis based on Segment
df.groupby("Segment")["Sales"].sum().plot.line()
```

```python
##Total sales of each Category
df.groupby("Category")["Sales"].sum()
```

```python
##sales analysis based on Category
df.groupby("Category")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

```python
##sales analysis based on State
df.groupby("State")["Sales"].sum().plot.bar()
```

  **4.2 Profit analysis**


```python
##Total profit of each Region
df.groupby("Region")["Profit"].sum()
```

```python
##profit analysis based on region
df.groupby("Region")["Profit"].sum().plot.bar()
```

```python
##Total profit of each Segment
df.groupby("Segment")["Profit"].sum()
```

```python
##profit analysis based on Segment
df.groupby("Segment")["Profit"].sum().plot.line()
```

```python
##Total profit analysis of each Category
df.groupby("Category")["Profit"].sum()
```

```python
##profit analysis based on Category
df.groupby("Category")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
##profit analysis based on state
df.groupby("State")["Profit"].sum().plot.bar()
```

  **5.Visualization of data**

```python
##Bar plot between Sub-category vs Category

plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=df)
plt.show()
```

**Bar plot of Sales vs Profit in Sub-Category**

```python
##Bar plot of Sales vs Profit
df.groupby('Sub-Category')[['Profit', 'Sales']].agg(['sum']). plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.rcParams['figure.figsize'] = [10,8]
plt.show()
```

**Sales and Profit Region wise**

```python
##Sales and Profit Region wise
df.groupby('Region')[['Profit', 'Sales']].agg(['sum']). plot.bar()
plt.title('Total Profit and Sales per Region')
plt.rcParams['figure.figsize'] = [10,8]
plt.show()
```
